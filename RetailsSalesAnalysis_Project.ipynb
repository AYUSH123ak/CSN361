```python
# 1.Read Data : Load the csv data into a Pandas DataFrame.
# 2.Data Cleaning:
#       Check for and remove rows with missing values in any columns.
# 3. Data Manipulation and Analysis:
#       Calculate total sales per product.
#       Identify the best-selling product.
#       Compute average daily sales.
# 4.Visualization:
#       Plot sales trends overtime.
#       Display sales pper product in a bar chart.
# 5.Class and Functions:
#       Define a Pthon class RetailsSalesAnalyzer with methods for each of the above tasks(loading data,Calculating statistics and creating visualizations)
# 6.Svcript Execution:
#       Write a script that creates an instance of retailsaslesanalyzer, calls its methods,and siplay the results.


import matplotlib.pyplot as plt
import pandas as pd
class RetailSalesAnalyzer:
  def __init__(self):
    self.data=pd.read_csv('/content/retail_sales.csv')
    self.data['Date']=pd.to_datetime(self.data['Date'])

  def data_cleaning(self):
    self.data.dropna(inplace=True)

  def total_sales_per_product(self):
    return self.data.groupby('Product')['Sales'].sum()

  def best_sell_product(self):
    return self.total_sales_per_product().sort_values(ascending=False).index[0]

  def average_sales(self):
    return self.data['Sales'].mean()


  def plot_trend(self):
    self.data.groupby('Date')['Sales'].sum().plot(kind='line')
    plt.title('Sales overtrend')
    plt.xlabel('Date')
    plt.ylabel('Total Sales')
    plt.show()

  def plot_bar_graph(self):
    self.total_sales_per_product().plot(kind='bar')
    plt.title('Sales over trend on bar chart')
    plt.xlabel('Date')
    plt.ylabel('Total sales')
    plt.show()
```

```python
analyzer=RetailSalesAnalyzer()
print('Total Sales per product: \n', analyzer.total_sales_per_product())
print('Best selling product: \n', analyzer.best_sell_product())
print('Average daily sales: \n', analyzer.average_sales())
analyzer.plot_trend()
analyzer.plot_bar_graph()
```
