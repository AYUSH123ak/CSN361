Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Read the Netflix dataset

```python
df=pd.read_csv('/content/netflix_titles.csv')
df.head()
```

Check whether their is null values present in the dataset or not

```python
df.isnull().sum()
```

Drop Null Values.

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.describe()
```

Convert date_added column to numeric value

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df.head()
```

Movies vs TV shows

```python
plt.figure(figsize=(5,5))
sns.countplot(x=df['type'],color='skyblue')
plt.title('Movies vs TV Show')
plt.xlabel('Type')
plt.show()
```

```python
country_counts = df['country'].value_counts().head(10)
country_counts
```

Top 10 countries watching Netflix.

```python
country_counts.plot(kind='bar',figsize=(10,6))
plt.title('Top 10 Countries')
plt.xlabel('Country')
plt.show()
```

```python
df.head()
```

Top 10 most categories wacthed in Netflix

```python
top_categories = df['listed_in'].str.split(', ').explode().value_counts().head(10)
top_categories.plot(kind='bar',color='skyblue',figsize=(10,6))
plt.title('Top 10 most categories watched in Netflix')
plt.xlabel('Category')
plt.ylabel('Number of titles')
plt.show()
```

Ratings of Tv shows and Movies

```python
plt.figure(figsize=(10,5))
sns.countplot(x=df['rating'],color='skyblue')
plt.title('Rating of tv shows and Movies')
plt.xlabel('Rating')
plt.ylabel('Number of titles')
plt.show()
```

Top 10 directors.

```python
Top_directors=df['director'].value_counts().head(10)
Top_directors.plot(kind='bar',color='brown',figsize=(10,6))
plt.title('Top 10 directors')
plt.xlabel('Director name')
plt.ylabel('Count')
plt.show()
```

```python
Top_countries=df['country'].value_counts().sort_values(ascending=False).head(10)
Top_countries
```

Top 10 countries watching Netflix

```python
Top_countries.plot(kind='pie',color='red',figsize=(10,6),startangle=90,shadow=True)
plt.legend(title='Top 10 Countries')
plt.title('Top 10 Countries watching Netflix')
plt.show()
```
